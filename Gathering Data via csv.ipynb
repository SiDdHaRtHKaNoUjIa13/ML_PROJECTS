# Convertors

```python
def rename(name):
    if name=='Mumbai Indians':
        return "MI"
    else:
        return name
```

```python
rename('Mumbai Indians')
```

```python
# we apply the function to the table and we can convert all the names of mumbai indians as mi

pd.read_csv('Ipl Matches 2008-2020.csv',converters = {'team1': rename})
```
